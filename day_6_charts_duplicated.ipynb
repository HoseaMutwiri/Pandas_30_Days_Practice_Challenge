```python
# Import the Pandas library
import pandas as pd
```

```python
#1. Displaying a custom number of records (tail)
import pprint

# Loading the dataset from the provided URL
def load_data(url):
    chunks = pd.read_csv(url,chunksize = 50)
    data = pd.concat(chunks)
    shape_dim = data.shape
    columns = list(data.columns)
    datatypes = data.dtypes.to_dict()
    print(f"Data loaded its shape : {shape_dim}")
    return data

data_url = "https://afterwork.ai/ds/e/information_technology_wign0.csv"

df_data = load_data(data_url)

# Displaying the last 5 records from the DataFrame
df_data.tail(5)
```

```python
# Sorting the dataset by multiple columns: Department (ascending) and Salary (descending)

df_data.sort_values(by=["Department","Salary"],ascending = [True,False]).head(5)
```

```python
df_data["Salary"].dtypes
```

```python
#3. Data type conversion using astype()
df_data["Salary"] = df_data["Salary"].astype('float')
print(df_data["Salary"].dtype)
```

```python
data3_url ="https://archive.org/download/information_technology_yecwl/information_technology_yecwl.csv"

data3 = pd.read_csv(data3_url)

# Check for missing values

data3.isnull().sum()
```

```python
# Calculate the mode of the column

mode_value = data3["Salary"].mode()[0]
mode_value
# Replace missing values with the mode value
data3.fillna(mode_value,inplace = True)
```

```python
# Check for missing values imputed using mode

nulld = data3[data3.isna().any(axis = 1)]
nulld
```

```python
# Loading the dataset from the provided URL
data4 = pd.read_csv('https://afterwork.ai/ds/ch/operations_x56c0.csv')
data4.head()
```

```python
# Ranking the data based on the 'Total Cost' column in descending order
data4["ranked"] = data4["Total Cost"].rank(method = "max",ascending =False)

# Displaying the top 3 most expensive products along with their 'Product Name' and 'Supplier'
data4[["Product Name","Supplier","Total Cost","ranked"]].sort_values(by = ["ranked"], ascending = [True]).head(3)
```
