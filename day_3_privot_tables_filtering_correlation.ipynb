```python
# Loading the dataset from the provided URL
boolean_indexing_data = pd.read_csv('https://afterwork.ai/ds/e/education_pkagx.csv')
boolean_indexing_data.head()
```

```python
# Applying Boolean Indexing to filter students who passed in Math with a score greater than or equal to 90

# creating a mask

mask = boolean_indexing_data["Math Score"]>=90

# filter the data using the mask

filtered_data = boolean_indexing_data[mask]

# display the filtered data
filtered_data.head()
```

```python
# Remove duplicate rows based on all columns

removed_duplicate_data = boolean_indexing_data.drop_duplicates()

removed_duplicate_data.head()
```

```python
# Loading the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/students_m6w10.csv')
df.head()
```

```python
# Load the dataset from the URL
df = pd.read_csv('https://afterwork.ai/ds/ch/schools_q9u1v.csv')
df.head()
```
