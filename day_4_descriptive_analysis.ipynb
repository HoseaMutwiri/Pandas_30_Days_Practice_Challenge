```python
# Import pandas for data manipulation
import pandas as pd
```

```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/banking_0bz3y.csv')
print(df.head())

# Grouping the data by 'Branch' and calculating the total 'Amount' for each branch

grouped_by_branch = df.groupby("Branch")["Amount"].sum()

# Displaying the total amount for each branch
grouped_by_branch
```

```python
# returns the columns
print(df.columns)

# Grouping the data by 'Transaction Type' and calculating the total 'Amount' for each transaction type

grouped_by_transaction_type = df.groupby("Transaction Type")["Amount"].sum()

# Displaying the total amount for each transaction type
print(grouped_by_transaction_type)
```

```python
pivot_table_1 = df.pivot_table(index=["Transaction Type","Card Type","Branch","Account Number"],values=["Amount"],aggfunc="sum")

print(pivot_table_1)
```

```python
# Selecting a range of rows and columns by label for data transformation

#method 1

df1 = df.iloc[:11,0:3]

print(df1)
print("====="*15)
#method 2

df2 =df.loc[:10,['Customer ID','Account Number','Transaction Type']]

print(df2)
```

```python
#Arbitrary Imputation

# Check for missing values

missing_values_check = df.isnull().sum()

print(missing_values_check)
```

```python
# Perform Arbitrary Imputation by replacing missing values with a specific value (e.g., 0)

df[["ATM Location","Merchant Name"]]= df[["ATM Location","Merchant Name"]].fillna("Not provided")

not_provided = df[df["Merchant Name"].str.contains("Not provided")|df["ATM Location"].str.contains("Not provided")]

print(not_provided)

missing = df.isnull().sum()
print(missing)
```

```python
#Filtering using the Query Method

filtered_data = df.query("`ATM Location` == 'Not provided'")
filtered_data.head()
```

```python

# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/banking_97kb8.csv')


#Encoding Data using the Map Function
# Define a mapping dictionary for encoding Account Type to numeric values
gender_map = {"Male":0,"Female":1}

# Apply the map function to encode Account Type column using the mapping dictionary
df["Gender_encoded"] = df["Gender"].map(gender_map)

df.head()
```

```python
# Define a mapping dictionary for encoding Account Type to numeric values

loan_status_map = {"Approved":1,
                   "Rejected":0}


# Apply the map function to encode Account Type column using the mapping dictionary

df["loan_status_encoded"] = df["Loan Status"].map(loan_status_map)

# Preview the resulting dataframe

df.head()
```

```python
#Random Sampling for Data Analysis
# Rows with a higher value in the "loan_status_encoded" column are more likely to be picked
# random_state =1 :ensure that the exact same 5 rows are chosen every single time you run the code.


sample_data = df.sample(n=5,weights="loan_status_encoded",random_state=1)
sample_data
```
