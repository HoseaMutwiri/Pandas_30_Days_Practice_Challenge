**Challenge**


Given the dataset from the URL: https://afterwork.ai/ds/ch/operations_wsg5u.csv, create a new column named 'Total Discount' in the DataFrame. The 'Total Discount' column should represent the total discount amount for each order, calculated by multiplying the 'Quantity' with the 'Unit Price' and then multiplying the result by the 'Discount' percentage. Use the 'assign()' method in Pandas to add this new column.

```python
# Load the dataset from the URL
df = pd.read_csv('https://afterwork.ai/ds/ch/operations_wsg5u.csv')
df.head()
```

```python
# Create a new column 'Total Discount' by multiplying 'Quantity' with 'Unit Price' and then multiplying by 'Discount' percentage

df = df.assign(total_discount = (df['Quantity']*df['Unit Price'])*df['Discount'])

df.head()
```
