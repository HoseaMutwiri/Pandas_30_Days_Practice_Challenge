```python
#3. Discretization and Binning
# Load the dataset from the provided URL
data = pd.read_csv('https://afterwork.ai/ds/ch/operations_872m6.csv')
data["Income"].describe()
```

```python
# Define bins for discretization based on income distribution

# Create a new column 'Income Group' by binning the 'Income' column
bins = [0,35000,89000,100000]

labels =["Low Income","Medium Income","High Income"]

data["Income category"] = pd.cut(data["Income"],bins=bins, labels =labels)

# Display the updated dataset with the new 'Income Group' column
data.head()
```

```python
#4. Outlier Detection using Z-scores
# z-score = (i-mean)/std

data['Spending Score'].plot(kind='box', figsize=(10, 6), showfliers=True);

ss_mean = data["Spending Score"].mean()

ss_std = data["Spending Score"].std()

# calculate z-score

data["z score"] = (data["Spending Score"]-ss_mean)/ss_std

data["z score"].min()


```
