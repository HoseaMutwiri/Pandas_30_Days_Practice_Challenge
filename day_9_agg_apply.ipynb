**CHALLENGE 3**

Create a summary table by grouping the agriculture dataset based on the 'Crop Type' column and calculate the average 'Yield' for each crop type. Use the dataset from the URL: https://afterwork.ai/ds/ch/agriculture_954ch.csv

```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/agriculture_954ch.csv')
df.head()
```

```python
def group_by_crop_type(group):
    return group["Yield"].mean()

new_df = df.groupby("Crop Type").apply(group_by_crop_type).reset_index()

new_df.to_excel(f"grouped_data_{today}.xlsx",index = False)
```

**CHALLENGE 4**

Given the dataset of farmers with information on their planting and harvest dates, write a Python function to calculate the number of days between the planting date and the harvest date for each farmer. Use the dataset from the URL: https://afterwork.ai/ds/ch/farmers_sxqrw.csv


```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/agriculture_rwzfy.csv')
df.head()
```

```python
from datetime import datetime as dt
import pandas as pd

# Calculate the number of days between planting and harvest dates for each farmer

def calc_days(group):
    prant_date = pd.to_datetime(group["Planting Date"])
    harvest_date = pd.to_datetime(group["Harvest Date"])
    group["days"] = (harvest_date - prant_date).dt.days
    return group

# Displaying the updated DataFrame with the number of days between planting and harvest dates
df = calc_days(df)
df.head()
```
