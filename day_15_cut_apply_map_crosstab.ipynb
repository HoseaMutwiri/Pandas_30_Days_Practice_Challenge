**CHALLENGE 4**

Using the dataset from the URL: https://afterwork.ai/ds/ch/sales_gc6tu.csv, create a new categorical column in the DataFrame by binning the 'Total Price' data into three bins: 'Low', 'Medium', and 'High'. Analyze the sales data based on these price categories. Remember to specify the data to be binned, define the bin edges or categories, and use pd.cut to create the new categorical column.

```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/sales_tn8b1.csv')

# create bins

dbins_labels = ['Low','Medium','High']

dbins= list(range(0,1201,400))

print(dbins)

df["binned_prices"] =pd.cut(df["Total Price"],bins = dbins,labels = dbins_labels,include_lowest=True)

df.head()
```

**CHALLENGE 6**

Given the dataset of products available at the URL: https://afterwork.ai/ds/ch/products_v9oui.csv, create a function that calculates the total revenue generated by each product by multiplying the 'Price' and 'Stock Quantity' attributes. Apply this function to every row in the dataset and store the results in a new column named 'Total Revenue'.

```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/products_v9oui.csv')
df.head()
```

```python
def total_revenue(group):
    revenue = group["Price"]*group["Stock Quantity"]
    return revenue.sum()


summary = df.groupby('Product Name').apply(total_revenue)

summ = pd.DataFrame(summary)

summ.T

```
