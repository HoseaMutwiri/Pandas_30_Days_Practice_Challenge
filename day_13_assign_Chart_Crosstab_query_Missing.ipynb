**Challange 1**

Using the dataset from the following URL: https://afterwork.ai/ds/ch/operations_ms39v.csv, Creating a new column 'Discounted Price' by applying a discount of 10% on 'Total Price'

```python
# Loading the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/operations_ms39v.csv')
df = df.assign(Discounted_Price= (df["Total Price"]*0.1))

df.head()
```

```python
q1 = df["Unit Price"].quantile(0.25)
q3 = df["Unit Price"].quantile(0.75)

iqr = q3-q1

lower = q1 -(1.5*iqr)

upper = q1 + (1.5*iqr)

rows,columns = df[(df['Total Price'] > upper) | (df['Total Price'] < lower)].shape

print(f"we have {rows} anomalities record")
```

**Challenge 5**

Given the HR dataset available at the URL: https://archive.org/download/hr_apxb7/hr_apxb7.csv, write a Python script using Pandas to handle missing data in the 'Age' column. You need to replace the missing values in the 'Age' column with the median age of all employees. Make sure to load the dataset and display the 'Age' column before and after handling missing data.

```python
# Load the dataset from the provided URL
data = pd.read_csv('https://archive.org/download/hr_apxb7/hr_apxb7.csv')
data.head()
```

```python
# Check for missing values

print("Null values before: " ,data["Age"].isnull().sum())

# Handling missing data by replacing missing values in 'Age' column with median age
median = data["Age"].median()

print("Median: ",median)

data["Age"].fillna(median,inplace=True)



# Confirming missing values
print("Null values after: " ,data["Age"].isnull().sum())
```
