**Challenge 5**

Given the dataset provided at the URL: https://afterwork.ai/ds/ch/operations_yux8o.csv, write a Python script using Pandas to search for all rows where the 'Delivery Status' column contains the value 'In Transit'. Once you have filtered the dataset, calculate the total quantity of products that are currently 'In Transit'.

```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/operations_yux8o.csv')

# Previw the dataset to understand its structure
df.head()
```

```python
# Search for all rows where 'Delivery Status' column contains 'In Transit'

filtered_data = df[df["Delivery Status"].str.contains("In Transit", case = False)]

# Calculate the total quantity of products that are currently 'In Transit'
# Display the total quantity of products that are currently 'In Transit'

print(filtered_data['Quantity'].agg(['sum']))
```
