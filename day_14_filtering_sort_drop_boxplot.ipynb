**CHALLENGE 1**

Given the dataset from the URL: https://afterwork.ai/ds/ch/factory_imclo.csv, create a code snippet using filtering rows with loc to extract factories that have a production capacity greater than 10,000. Display the Factory ID, Factory Name, Location, and Production Capacity of the filtered factories.




```python
# Loading the dataset from the provided URL

factory = pd.read_csv("https://afterwork.ai/ds/ch/factory_imclo.csv")

factory.head()
```

```python
# Filtering rows using loc based on a condition (e.g., Production Capacity greater than 10,000)

filtered = factory.loc[factory["Production Capacity"]>10000,["Factory ID", "Factory Name", "Location", "Production Capacity"]]

# Previewing the filtered factories
filtered.head()
```

**CHALLENGE 2**

Given the dataset from the URL: https://afterwork.ai/ds/ch/factory_obq2k.csv, create a code snippet using the Pandas query method to filter out all factories where the Production Capacity is greater than 10,000 and the Total Employees are greater than 200.


```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/factory_obq2k.csv')
df.head()
```

```python
# Apply advanced filtering using the query method

a_filter = df[["Factory ID", "Factory Name", "Location", "Production Capacity","Total Employees"]].query("`Production Capacity`>10000 and `Total Employees` >200")
a_filter.head()
```

**CHALLENGE 4**

Given the dataset of factories available at the URL: https://afterwork.ai/ds/ch/factory_7gsiv.csv, create a code snippet using Pandas to sort the factories first by 'Production Capacity' in descending order and then by 'Total Employees' in ascending order. This will help in identifying the factories with the highest production capacity and the least number of employees within each capacity group.

```python
# Loading the dataset from the provided URL
df_4 = pd.read_csv('https://afterwork.ai/ds/ch/factory_7gsiv.csv')
df_4.head()
```

```python
# Sorting the DataFrame by multiple columns 'Production Capacity' in descending order and 'Total Employees' in ascending order

sorted = df_4.sort_values(by = ["Production Capacity","Total Employees"], ascending = [False,True])


# Previewing the sorted DataFrame
sorted.head()
```
