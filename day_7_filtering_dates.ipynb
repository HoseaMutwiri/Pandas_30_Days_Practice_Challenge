```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/operations_aw02v.csv')
df.head()
```

```python
# Define a custom function to calculate discounted price

def calculate_discounted_price(dt):
    price = dt["Price"]
    Quantity = dt["Quantity"]
    Discount = dt["Discount"]
    return (price*Quantity)-Discount

#Apply the custom function to create a new column 'Discounted Price'
df["price_with_discount"] = df.apply(calculate_discounted_price,axis= 1)

df.head()
```

```python
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/hr_kix46.csv')
df.head()
```

```python

# Convert 'Joining Date' column to datetime format
df['Joining Date'] = pd.to_datetime(df['Joining Date'])

# Set 'Joining Date' as the index of the DataFrame
df.set_index('Joining Date', inplace=True)

# Sort the DataFrame by the index to ensure it is monotonic
df.sort_index(inplace=True)

# Perform time-based slicing to extract data for a specific time period
start_date = '2018-01-01'
end_date = '2019-12-31'
time_period_data = df[start_date:end_date]

# Preview the extracted data
time_period_data.head()
```

```python
# Converting 'Joining Date' column to datetime format
df['Joining Date'] = pd.to_datetime(df['Joining Date'])

# Filtering data based on a specific date range (e.g., from 2019-12-01 to 2020-02-29)
filtered_data = df[df['Joining Date'].between('2019-12-01', '2020-02-29')]

# Displaying the filtered data
filtered_data.head()
```

```python
# Creating a summary table using pivottable() function
summary_table = pd.pivot_table(df, index=['Department'], values='Salary', aggfunc='mean').reset_index()

# Displaying the summary table
summary_table
```

```python
# Converting 'Joining Date' column to datetime format
df['Joining Date'] = pd.to_datetime(df['Joining Date'])

# Sorting the records by the 'Joining Date' column in Ascending order
df = df.sort_values('Joining Date')

# Setting the 'Joining Date' column as the index column
df.set_index('Joining Date', inplace=True)

# Create a line plot
df['Salary'].plot(kind='line', title='Salary Trend Over Time');
```

```python
data = pd.read_csv('https://afterwork.ai/ds/e/healthcare_and_pharmaceuticals_zulr.csv')

missing_values = data.isnull().sum()
print(missing_values)


#----------------------------------------------------------------------------------------

from sklearn.impute import KNNImputer

# Initialize the KNNImputer with k=5 (number of neighbors to consider)
imputer = KNNImputer(n_neighbors=5)

# Perform KNN Imputation on the 'Age' column (assuming 'Age' has missing values)
data['Age'] = imputer.fit_transform(data[['Age']])


# Check for missing values
missing_values = data.isnull().sum()
missing_values
```
