```python
#Handling Missing Data in Pandas

# Load the dataset from the provided URL
df1 = pd.read_csv('https://afterwork.ai/ds/e/healthcare_nqob7.csv')

# Display the dataset to check for missing values
df1.isnull().sum()
```

```python
# Fill missing values in the 'Weight' column with the mean value of the column

df1["Weight"].fillna(df1["Weight"].mean())

df1.isnull().sum()
```

```python
# Drop rows with any missing values in the dataset
df1.dropna(axis=0,how="any",inplace=True)
df1.isnull().sum()
```

```python
#Date Range Filtering
# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/ch/healthcare_u75x4.csv')
df.head()

```

```python
# Convert 'Admission Date' column to datetime format

df["Admission Date"] = pd.to_datetime(df["Admission Date"])
#verify the data type
print(df["Admission Date"].dtype)

df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])

print(df.dtypes)
```

```python
# Apply date range filtering to focus on data within the specified range

df[((df["Discharge Date"] - df["Admission Date"])>pd.Timedelta(weeks=3)) & (df["Gender"]=="Male")]
```
