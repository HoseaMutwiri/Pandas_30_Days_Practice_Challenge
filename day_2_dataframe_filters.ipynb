```python
# Importing the Pandas library
import pandas as pd

# Loading the dataset from the provided URL

df = pd.read_csv('https://afterwork.ai/ds/e/healthcare_and_pharmaceuticals_hjvuy.csv')

# Selecting specific data columns for analysis
# 'Patient Name', 'Age', 'Diagnosis', 'Treatment', 'Medication', 'Doctor'

print(df.head())

new_data = df[['Patient Name', 'Age', 'Diagnosis', 'Treatment', 'Medication', 'Doctor']]

# Displaying the selected data
new_data.head()
```

```python
cleaned_data = df.dropna().drop_duplicates() # removed duplicates and null values
# confrim
cleaned_data.isnull().sum()
```

```python
cleaned_data.head()
```

```python
# Convert 'Admission Date' column to datetime format
# check the data type
cleaned_data.dtypes

cleaned_data["Admission Date"] = pd.to_datetime(cleaned_data["Admission Date"])

cleaned_data.dtypes
```

```python
# Grouping the data by 'Diagnosis' and calculating the average age for each diagnosis

cleaned_data.columns

grouped_data_by_diagnosis = cleaned_data.groupby("Diagnosis")["Age"].mean()

grouped_data_by_diagnosis
```

```python
# Grouping the data by 'Gender' and counting the number of patients for each gender

grouped_by_gender_value_patients = cleaned_data.groupby("Gender")["Patient ID"].count()

grouped_by_gender_value_patients
```

```python
# Filter the dataset based on multiple conditions using logical AND operator

# Load the dataset from the provided URL
df = pd.read_csv('https://afterwork.ai/ds/e/healthcare_1xeq5.csv')
df.head()

filtred_data = df[(df["Age"]<=30) & (df["Diagnosis"]== "Diabetes")]

filtred_data
```

```python
# Negating Filters with NOT

filtred_data = df[~(df["Diagnosis"]== "Diabetes") & ~(df["Gender"]=="Female")]

filtred_data
```

```python
# Filter out rows where the 'Diagnosis' column is not equal to 'Diabetes'

filtred_data = df[~(df["Diagnosis"]== "Diabetes")]

filtred_data
```

```python
# Filter the DataFrame to select rows where the patient's age is greater than 50 or the patient's diagnosis is 'Cancer'

filtred_data_2 = df[(df["Age"]>50) | (df["Diagnosis"]=="Cancer")]
filtred_data_2
```

```python

#String Extraction Filtering

string_filtered = df[df["Treatment"].str.contains("Therapy")]

string_filtered

```

```python
# Filter the dataset to extract only the patients with a specific diagnosis, for example, 'Depression'

filtered_depression = df[df["Diagnosis"].str.contains("Depression")]

filtered_depression
```

```python
#Date Range Filtering

# Convert 'Admission Date' and 'Discharge Date' columns to datetime format
df['Admission Date'] = pd.to_datetime(df['Admission Date'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])

# Define the date range for filtering
start_date = '2021-08-01'
end_date = '2022-08-31'


filtered_by_date = df.query("`Admission Date`>=@start_date and `Discharge Date`<=@end_date")

filtered_by_date
```
